üìå we will use *get_mood* function, to determine the mood of a given text using sentiment analysis

üìå the function utilizes the *SentimentIntensityAnalyzer* from vaderSentiment to calculate the sentiment score of the unput text, then compares the score with the threshold

```python
from dataclasses import dataclass
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

@dataclass
class Mood:
    emoji: str
    sentiment: float

    @staticmethod
    def get_mood(input_text: str, *, threshold: float = 0.01) -> 'Mood':
        analyzer = SentimentIntensityAnalyzer()
        sentiment_dict = analyzer.polarity_scores(input_text)
        sentiment = sentiment_dict['compound']

        friendly_threshold: float = threshold
        hostile_threshold: float = -threshold

        if sentiment >= friendly_threshold:
            return Mood("üòä", sentiment)
        elif sentiment <= hostile_threshold:
            return Mood("üò°", sentiment)
        else:
            return Mood("üòê", sentiment)
```

üìå creating a new column contains the mood of each comment based on its sentiment, using get mood function

```python
df['mood'] = df['clean_comment'].astype(str).apply(Mood.get_mood) # applying get mood function on string only
```

üìå next, we will create two more columns, using lambda function we will extract the emoji and the sentiment score for each comment, to make it easier to see the sentiment of each comment

```python
df['emoji'] = df['mood'].apply(lambda x: x.emoji)
df['sentiment'] = df['mood'].apply(lambda x: x.sentiment)
```
